```python
pathname = 'https://raw.githubusercontent.com/osoliman/Chapter6_HTM737/refs/heads/main/' #isolated cause it is in all file names
files_of_interest = [
    'hvbp_tps_11_07_2017.csv',
    'hvbp_clinical_care_11_07_2017.csv',
    'hvbp_safety_11_07_2017.csv',
    'hvbp_efficiency_11_07_2017.csv',
    'hvbp_hcahps_11_07_2017.csv'
]#list due to square brackets

dfs = {
    foi: pd.read_csv(pathname + foi, header=0) for foi in files_of_interest #curly brackets indicate dictionary
}
```

```python

for df in dfs.values():
    df.rename(columns={'Provider_Number': 'Provider Number'}, inplace=True)
```

```python

df_master = dfs[files_of_interest[0]]

for foi in files_of_interest[1:]:
    df_master = df_master.merge(
        dfs[foi],
        on='Provider Number',
        how='left',
        copy=False
    )

print(df_master.shape) #error given after 3 repetitions
```

```python
from collections import Counter


all_columns = []
for df in dfs.values():
    all_columns.extend(df.columns)


column_counts = Counter(all_columns)


common_cols = {col for col, count in column_counts.items()
               if count > 1 and col != 'Provider Number'}

print(f"Columns to drop: {common_cols}") #columns repeated are lsited below
```

```python
for col, count in column_counts.most_common(): #extra stepp to doublecheck
    if count > 1:
        print(f"{col}: {count} times")
```

```python

df_master = dfs[files_of_interest[0]]

for foi in files_of_interest[1:]:
    df_to_merge = dfs[foi].drop(columns=common_cols, errors='ignore')
    df_master = df_master.merge(
        df_to_merge,
        on='Provider Number',
        how='left',
        copy=False
    )

print(df_master.shape)
```

```python
df_master.to_csv('hvbp.csv', index=False) #writes cvs while ignoring first row index numbers
```
